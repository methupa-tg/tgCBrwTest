```python
import cohere
import os

api_key = os.getenv("COHERE_API_KEY")
if not api_key:
    raise EnvironmentError("COHERE_API_KEY is not set. Add it to your environment variables.")

co = cohere.ClientV2(api_key)
```

```python
prompt = "You are the Customer Support Assistant of Thyaga.lk. " \
         "Recommend vouchers to users based on their preferences. " \
         "Answer questions about Thyaga.lk. " \
         "Keep the answers short and simple. " \
         "Do not answer any other questions which are not related to thyaga vouchers or thyaga basic information. " \
         "If the user asks any question which is not related to thyaga vouchers or thyaga basic information, answer with 'Sorry, I can only answer questions related to Thyaga vouchers and basic information.'"

chat_history = [{'role': 'system', 'content': prompt}]

print("Thyaga Assistant (type 'quit' to exit)\n")

while True:
    message = input("You: ").strip()
    if message.lower() == 'quit':
        break
    if not message:
        continue

    chat_history.append({'role': 'user', 'content': message})

    response = co.chat_stream(
        model="command-a-03-2025",
        messages=chat_history,
        temperature=0.25,
        max_tokens=300,
        frequency_penalty=0.4
    )

    print("Assistant: ", end='', flush=True)
    full_response = ""
    for event in response:
        if event.type == 'content-delta':
            text = event.delta.message.content.text
            print(text, end='', flush=True)
            full_response += text
    print()

    chat_history.append({'role': 'assistant', 'content': full_response})
```

```python
for message in chat_history:
    role = message['role'].capitalize()
    content = message['content']
    print(f"{role}: {content}\n")
```
